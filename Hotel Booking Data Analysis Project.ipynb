# data analysis and visualization

```python
len(df1[df1.is_canceled == 1])
```

```python
cancel_pre = df1['is_canceled' ].value_counts(normalize = True)
print(cancel_pre)
```

```python
cancel_pre = df1['is_canceled' ].value_counts(normalize = True)
print(cancel_pre)

plt.figure(figsize =( 5,4))
plt.title( "reservation_status")
plt.bar(['not cancel' , 'canceled'] , df1['is_canceled' ].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8,4))
axl = sns.countplot(x = "hotel" , hue = "is_canceled"  , data= df1 , palette = 'Blues' ) 
legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation_status_in_different_hotels')
plt.xlabel('hotels')
plt.ylabel('number_of_reservation')
```

```python
cancellation_counts = df1.groupby(['hotel', 'is_canceled']).size().unstack()
cancellation_counts
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = ( 20,8))
plt.title('average dailly rate in city and resort hotel' , fontsize = 20)
plt.plot(resort_hotel.index , resort_hotel['adr'] , label = 'resort_hotel')
plt.plot(city_hotel.index , city_hotel['adr'] , label = 'city_hotel')
plt.legend(fontsize = 20)
```

```python
df1['months'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
axl = sns.countplot(x = 'months' , hue = 'is_canceled' , data = df1)
legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation_status_per_month' , size = 20)
plt.xlabel('months')
plt.ylabel('number_of_reservation')
plt.legend(['not cancel' , 'canceled'])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per months' , fontsize = 23)
axl = sns.barplot('months' ,'adr', data = df1[df1['is_canceled']==1].groupby('months')[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('top_10_country with reservation canceled')
plt.pie(top_10_country , autopct = '%0.2f' , labels = top_10_country.index)
plt.show()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date'  , inplace = True)

not_canceled_df_adr =  df[df['is_canceled']==0]
not_canceled_df_adr = not_canceled_df_adr.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date'  , inplace = True)

plt.figure(figsize = (20,10))
plt.title('Average daily rate')
plt.plot(not_canceled_df_adr['reservation_status_date'] , not_canceled_df_adr['adr'] , label = 'not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'] ,canceled_df_adr['adr'] , label = 'cancelled')
plt.legend()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date'  , inplace = True)
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2015-07' ) & (canceled_df_adr['reservation_status_date'] <'2017-9')]

not_canceled_df_adr =  df[df['is_canceled']==0]
not_canceled_df_adr = not_canceled_df_adr.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date'  , inplace = True)
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] >'2015-07') & (not_canceled_df_adr['reservation_status_date']<'2017-9' )]

plt.figure(figsize = (20,10))
plt.title('Average daily rate')
plt.plot(not_canceled_df_adr['reservation_status_date'] , not_canceled_df_adr['adr'] , label = 'not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'] ,canceled_df_adr['adr'] , label = 'cancelled')
plt.legend()
```
